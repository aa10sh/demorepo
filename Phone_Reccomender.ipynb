```python
print(df_demo.duplicated().value_counts())
```

```python
df_demo.isnull().sum()
```

```python
display(df_demo[df_demo.isnull().any(axis=1)])
```

```python
df_demo.dropna(subset=['Battery'], inplace=True)
print(df_demo.shape)
```

```python
df_demo.isnull().sum()
```

```python
display(df_demo[df_demo['ROM/Storage'].isnull()])
```

```python
df_demo.dropna(subset=['ROM/Storage'], how='all', inplace=True)
print(df_demo.shape)
```

```python
df_demo.isnull().sum()
```

```python
df_demo.dropna(subset=['Front Camera'],how='all',inplace=True)
```

```python
print(df_demo.shape)
```

```python
df_demo.isnull().sum()
```

```python
df_demo.info()
```

```python
df_demo.rename(columns={'Phone Name': 'Phone','Rating ?/5': 'Rating','ROM/Storage':'ROM'}, inplace=True)
```

```python
print(df_demo.columns)
```

```python
df_demo.sample(5)
```

```python
# Convert 'Number of Ratings' to numeric
df_demo['Number of Ratings'] = df_demo['Number of Ratings'].str.replace(',', '').astype(int)

# Convert 'Price in INR' to numeric
df_demo['Price in INR'] = df_demo['Price in INR'].str.replace('₹', '').str.replace(',', '').astype(int)

# Calculate and display the correlation matrix
display(df_demo.corr(numeric_only=True))
```
